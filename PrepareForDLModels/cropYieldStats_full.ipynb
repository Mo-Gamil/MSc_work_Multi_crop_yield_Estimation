```python
import pandas as pd

def display_column_statistics(file_path: str, column1: str, column2: str):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Ensure the columns exist in the DataFrame
    if column1 not in df.columns or column2 not in df.columns:
        raise ValueError(f"One or both columns '{column1}' and '{column2}' do not exist in the CSV file.")
    
    # Calculate statistics for the specified columns
    stats_column1 = df[column1].describe()
    stats_column2 = df[column2].describe()
    
    # Display the statistics
    print(f"Statistics for {column1}:")
    print(stats_column1)
    
    print(f"\nStatistics for {column2}:")
    print(stats_column2)

import pandas as pd
import matplotlib.pyplot as plt

def display_column_statistics_and_graphs(file_path: str, column1: str, column2: str):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Ensure the columns exist in the DataFrame
    if column1 not in df.columns or column2 not in df.columns:
        raise ValueError(f"One or both columns '{column1}' and '{column2}' do not exist in the CSV file.")
    
    # Calculate statistics for the specified columns
    stats_column1 = df[column1].describe()
    stats_column2 = df[column2].describe()
    
    # Display the statistics
    print(f"Statistics for {column1}:")
    print(stats_column1)
    
    print(f"\nStatistics for {column2}:")
    print(stats_column2)
    
    # Plot graphs
    plt.figure(figsize=(14, 6))
    
    # Histogram for column1
    plt.subplot(2, 2, 1)
    plt.hist(df[column1].dropna(), bins=30, edgecolor='k')
    plt.title(f'Histogram of {column1}')
    plt.xlabel(column1)
    plt.ylabel('Frequency')
    
    # Box plot for column1
    plt.subplot(2, 2, 2)
    plt.boxplot(df[column1].dropna())
    plt.title(f'Box Plot of {column1}')
    plt.ylabel(column1)
    
    # Histogram for column2
    plt.subplot(2, 2, 3)
    plt.hist(df[column2].dropna(), bins=30, edgecolor='k')
    plt.title(f'Histogram of {column2}')
    plt.xlabel(column2)
    plt.ylabel('Frequency')
    
    # Box plot for column2
    plt.subplot(2, 2, 4)
    plt.boxplot(df[column2].dropna())
    plt.title(f'Box Plot of {column2}')
    plt.ylabel(column2)
    
    # Adjust layout
    plt.tight_layout()
    plt.show()


```
