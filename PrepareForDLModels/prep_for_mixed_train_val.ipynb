```python
import pandas as pd

# File paths
file_path_1 = r'ann_dir/Illinois_32616_2022_cropYield.csv'
file_path_2 = r'ann_dir/Iowa_32615_2022_cropYield.csv'
file_path_3 = r'ann_dir/Indiana_32615_2022_cropYield.csv'

output_file_path = r'ann_dir/IOWA_ILLINOIS_Ind_22_train_val.csv'

# Read the first CSV file
df1 = pd.read_csv(file_path_1)

# Read the second CSV file
df2 = pd.read_csv(file_path_2)

# Read the third Csv file
df3 = pd.read_csv(file_path_3)

# Concatenate the DataFrames
combined_df = pd.concat([df1, df2,df3], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv(output_file_path, index=False)

print(f"Combined CSV saved to {output_file_path}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the CSV file
file_path = 'ann_dir/IOWA_ILLINOIS_Ind_22_train_val.csv'
data = pd.read_csv(file_path)

# Split the data into 70% training and 30% testing
train_data, test_data = train_test_split(data, test_size=0.3, random_state=42)

# Save the splits into separate CSV files
train_data.to_csv('train_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)

print("Data has been split and saved to 'train_data.csv' and 'test_data.csv'.")
```

```python
import pandas as pd
import os
import shutil

# Read the CSV file
csv_file_path = 'train_data.csv'
data = pd.read_csv(csv_file_path)

# Extract the list of image names from the "Index" column
image_names = data['Index'].tolist()

# Define the source and destination folders
source_folder = 'img_dir/train'
destination_folder = 'img_dir/train_2'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Copy each image to the destination folder
for image_name in image_names:
    source_path = os.path.join(source_folder, f"{image_name}.tif")
    destination_path = os.path.join(destination_folder, f"{image_name}.tif")
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_path)
        print(f"Copied {source_path} to {destination_path}")
    else:
        print(f"Image {source_path} does not exist")

print("Image copying completed.")
```

```python
# Normalize training for training

# for training

import pandas as pd

# Load the CSV file
file_path = 'ann_dir/train_2/train_2_data.csv'  # Replace with the actual file path

df = pd.read_csv(file_path)

# Extract the columns
soy_column = df['soy_y_bu_per_patch']
corn_column = df['corn_y_bu_per_patch']

# Calculate min and max values for each column
soy_min = 0
soy_max = 71167.31490587837
corn_min = 0
corn_max = 243988.0697627311

# Print min and max values
print(f'Soy - Min: {soy_min}, Max: {soy_max}')
print(f'Corn - Min: {corn_min}, Max: {corn_max}')


# Normalize the columns
df['soy_y_normalized'] = (soy_column - soy_min) / (soy_max - soy_min)
df['corn_y_normalized'] = (corn_column - corn_min) / (corn_max - corn_min)


# Display the new dataframe
# import ace_tools as tools; tools.display_dataframe_to_user(name="Normalized DataFrame", dataframe=df)

# Save the dataframe if needed
df.to_csv('ann_dir/train_2/normalized_train_yield.csv', index=False)





```
