```python
for i, vars in enumerate(variable_sets):
    center_points_shapefile_path= vars['a'] 
    crs = vars['b'] 
    CDL_start_date = vars['c'] 
    CDL_end_date = vars['d'] 
    sen2_start_date = vars['e'] 
    sen2_end_date = vars['f']
    bands = ['B2', 'B3', 'B4','B5', 'B6','B7','B8',"B8A"]
    cloud_cover_max = 8
    buffer_dist = 1120 
    points = read_random_points(center_points_shapefile_path)
    

    # download cDL patches
    output_CDL_path= os.path.dirname(center_points_shapefile_path)
    output_CDL_path = os.path.join(output_CDL_path, 'CDL_images')
    os.makedirs(output_CDL_path, exist_ok=True)


    download_CDL_patches_from_GEE_to_local_desk(points,CDL_start_date,CDL_end_date,1120,output_CDL_path,crs)

    # download sent2 patches

    output_sen2_path= os.path.dirname(center_points_shapefile_path)
    output_sen2_path = os.path.join(output_sen2_path, 'sent2_images')
    os.makedirs(output_sen2_path, exist_ok=True)
    download_sent2_image_patches_from_GEE_to_local_desk(points,sen2_start_date,sen2_end_date,cloud_cover_max,1120,output_sen2_path,bands,crs)




    # clip sent2 patches
    #change size of sent2
    folder_path = output_sen2_path


    sent2_images_clipped_path = os.path.dirname(center_points_shapefile_path)
    sent2_images_clipped_path = os.path.join(sent2_images_clipped_path, 'sent2_images_clipped')
    os.makedirs(sent2_images_clipped_path, exist_ok=True)
    change_patches_size_in_folder(folder_path,sent2_images_clipped_path,224)

    #change size of cdl
    folder_path = output_CDL_path

    CDL_images_clipped_path = os.path.dirname(center_points_shapefile_path)
    CDL_images_clipped_path = os.path.join(CDL_images_clipped_path, 'CDL_images_clipped')

    os.makedirs(CDL_images_clipped_path, exist_ok=True)
    change_patches_size_in_folder(folder_path,CDL_images_clipped_path,224)


    # Normalize sent2 patches

    sent2_images_clipped_path = os.path.dirname(center_points_shapefile_path)
    sent2_images_clipped_path = os.path.join(sent2_images_clipped_path, 'sent2_images_clipped')
    os.makedirs(sent2_images_clipped_path, exist_ok=True)



    input_folder_before_normalization= sent2_images_clipped_path
    output_folder_after_normalization = os.path.dirname(center_points_shapefile_path)
    output_folder_after_normalization = os.path.join(output_folder_after_normalization, 'sent2_images_clipped_normalized')
    os.makedirs(output_folder_after_normalization, exist_ok=True)
    normalize_sent2_pixel_values(input_folder_before_normalization, output_folder_after_normalization)


    # change_background pixel values of cdl
    input_dir = CDL_images_clipped_path
    center_points_shapefile_path = center_points_shapefile_path
    CDL_images_final = os.path.dirname(center_points_shapefile_path)
    CDL_images_final = os.path.join(CDL_images_final, 'CDL_images_final')
    os.makedirs(CDL_images_clipped_path, exist_ok=True)
    setBackgroundToZero_of_CDL(input_dir, CDL_images_final)



    # Remap the pixel values of CDL
    # Specify the input and output folders
    input_folder = CDL_images_final
    CDL_images_final_mapped = os.path.dirname(center_points_shapefile_path)
    CDL_images_final_mapped = os.path.join(CDL_images_final_mapped, 'CDL_images_final_mapped')
    os.makedirs(CDL_images_final_mapped, exist_ok=True)
    

    # Process all images
    map_cdl_images_in_folder(input_folder, CDL_images_final_mapped)


    

    # calculate the crop yeild

    CDL_images_final_mapped = CDL_images_final_mapped
    points_shapefile_path = center_points_shapefile_path
    crop_yield = calculate_crop_yield(CDL_images_final_mapped,points_shapefile_path)


    # save the crop yield csv file


    # Extract directory and construct new CSV file path
    directory = os.path.dirname(center_points_shapefile_path)
    year = os.path.basename(directory)
    base_name = os.path.basename(os.path.dirname(directory))
    csv_file_name = f"{base_name}_{year}_cropYield.csv"
    csv_file_path = os.path.join(directory, csv_file_name)

    # Write list to CSV file
    with open(csv_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Index', 'patch_Name', 'soy_y_bu_per_m2', 'corn_y_bu_per_m2', 'soy_y_bu_per_patch', 'corn_y_bu_per_patch'])  # Write header
        for row in crop_yield:
            writer.writerow(row)  # Write each tuple as a row

        # print(f"List saved to {csv_file_path}")


```
